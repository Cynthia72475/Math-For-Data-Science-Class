```python
# ================================================================
# Title: Assignment 4.2 - Average Rate of Change
# Author: mikexcohen (GitHub username)
# Date: 1/7/2025
# Modified By: Cyndie Cramer
# Description: Create a Python function that takes in a 
# mathematical function, and two numbers a and b and returns 
# the average rate of change of the function between a and b.
# =================================================================

# Function to calculate the average rate of change
def average_rate_of_change(func, a, b):
    """
    Calculates the average rate of change of a function 'func'
    between two points a and b.
    
    Parameters:
    func -- a Python function representing the mathematical function
    a -- starting x-value
    b -- ending x-value
    
    Returns:
    The average rate of change (slope) between a and b
    """
    return (func(b) - func(a)) / (b - a)

# Example function: f(x) = 3x^2
def f(x):
    return 3 * x**2

# Define the interval for which we want to calculate the average rate of change
a = 1
b = 3

# Call the average_rate_of_change function and store the result
result = average_rate_of_change(f, a, b)

# Print the result
print(f"The average rate of change of f(x) = 3x^2 from x = {a} to x = {b} is {result}")
```
