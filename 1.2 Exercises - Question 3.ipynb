```python
# =====================================================
# Title: Assignment 1.2 
# Author: prathmachowksey (GitHub username)
# Date: 12/4/2025
# Modified By: Cyndie Cramer
# Description: This program calculates prediction 
# accuracy for binary labels for mushrooms
# ("p" = poisonous, "e" = edible).
# =====================================================

# import numpy
import numpy as np
import pandas as pd   # Pandas is used here to read the data file

# Set the path to your data file
file_path = r"C:\Users\cyndi\Documents\week1data\mushroom_data.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Convert the true and predicted price columns to NumPy
actual = df["actual"].to_numpy()
predicted = df["predicted"].to_numpy()

# Accuracy function
def accuracy_score(y_true, y_pred):
    """
    Calculates prediction accuracy:
    (# correct predictions) / (total predictions)
    """
    correct = np.sum(y_true == y_pred)
    total = len(y_true)
    return correct / total

# Compute accuracy
accuracy_value = accuracy_score(actual, predicted)

# Print result
print("Prediction Accuracy:", accuracy_value)
```
