```python
# =====================================================
# Title: Assignment 1.2 
# Author: rougier (GitHub username)
# Date: 12/4/2025
# Modified By: Cyndie Cramer
# Description: Plots the error function f(p) using 
# Matplotlib, # finds the p value that minimizes the 
# error, and zooms in around the minimum for clarity. 
# =====================================================

# import numpy
import numpy as np

# Import Matplotlib for creating plots and visualizations
import matplotlib.pyplot as plt   

# Define the function
def f(p):
    return (0.005*p**6 - 0.27*p**5 + 5.998*p**4 - 
            69.919*p**3 + 449.17*p**2 - 1499.7*p + 2028)

# Generate values for p
p_values = np.linspace(0, 20, 500)  # 500 points from 0 to 20
f_values = f(p_values)

# Find minimum error
min_index = np.argmin(f_values)   # index of minimum value
p_min = p_values[min_index]       # p value corresponding to minimum
f_min = f_values[min_index]       # minimum error

print("Minimum error occurs at p =", p_min)
print("Minimum error value =", f_min)

# Plot the function
plt.figure(figsize=(12,6))
plt.plot(p_values, f_values, color='blue', linewidth=2, label='f(p)')
plt.scatter(p_min, f_min, color='red', s=100, zorder=5, label=f'Minimum: f({p_min:.2f})={f_min:.2f}')

# Zoom in around the minimum
zoom_margin_p = 2  # p values around minimum
zoom_margin_f = 50  # f(p) values around minimum
plt.xlim(p_min - zoom_margin_p, p_min + zoom_margin_p)
plt.ylim(f_min - zoom_margin_f, f_min + zoom_margin_f)

plt.title("Error Function f(p) Zoomed Around Minimum", fontsize=16)
plt.xlabel("Parameter p", fontsize=14)
plt.ylabel("Error f(p)", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
