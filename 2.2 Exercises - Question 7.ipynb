```python
# =====================================================
# Title: Assignment 2.2 Question 6
# Author: sabnisshreyas91 (GitHub username)
# Date: 12/13/2025
# Modified By: Cyndie Cramer
# Description: Make a scatterplot of the highway miles 
# per gallon (y-axis) versus the weight (x-axis). Find 
# the root mean squared error (RMSE) of the prediction 
# line you found in problem (5) and the actual best-fit
# line found in problem (6).
# =====================================================

# Import the libraries we need
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

# Load the dataset using the full file path
data = pd.read_csv(r"C:\Users\cyndi\Documents\week2data\car_data.csv")

# Extract variables
x = data["weight"]
y = data["hwy_mpg"]

# --- Problem 5: Best-fit line using numpy.polyfit ---
slope_np, intercept_np = np.polyfit(x, y, 1)
y_pred_np = slope_np * x + intercept_np

# --- Problem 6: Best-fit line using scikit-learn ---
X = x.values.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)
y_pred_skl = model.predict(X)

# --- Calculate RMSE ---
rmse_np = np.sqrt(mean_squared_error(y, y_pred_np))
rmse_skl = np.sqrt(mean_squared_error(y, y_pred_skl))

print(f"RMSE (numpy.polyfit): {rmse_np:.4f}")
print(f"RMSE (scikit-learn LinearRegression): {rmse_skl:.4f}")
```
