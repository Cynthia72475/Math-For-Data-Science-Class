```python
# =====================================================
# Title: Assignment 2.2 Question 5
# Author: sabnisshreyas91 (GitHub username)
# Date: 12/13/2025
# Modified By: Cyndie Cramer
# Description: Make a scatterplot of the highway miles 
# per gallon (y-axis) versus the weight (x-axis). Add
# a best fit line to the plot and identify the slope
# and y-intercept.
# =====================================================

# Import the libraries we need
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset using the full file path
data = pd.read_csv(r"C:\Users\cyndi\Documents\week2data\car_data.csv")

# Extract the variables
x = data["weight"]
y = data["hwy_mpg"]

# Create the scatterplot
plt.scatter(x, y)

# Compute the line of best fit (y = mx + b)
slope, intercept = np.polyfit(x, y, 1)

# Create predicted y values using the best-fit line
y_fit = slope * x + intercept

# Plot the best-fit line
plt.plot(x, y_fit, label=f"y = {slope:.4f}x + {intercept:.2f}")

# Label the axes
plt.xlabel("Vehicle Weight (pounds)")
plt.ylabel("Highway Miles per Gallon")

# Add a title
plt.title("Highway MPG vs Vehicle Weight")

# Show the slope and intercept in a legend
plt.legend()

# Display the plot
plt.show()

# Print slope and intercept to the console
print(f"Slope: {slope}")
print(f"Y-intercept: {intercept}")
```
