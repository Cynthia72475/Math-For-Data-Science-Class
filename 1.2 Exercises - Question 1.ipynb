```python
# =====================================================
# Title: Assignment 1.2 
# Author: alifrmf (GitHub username)
# Date: 12/2/2025
# Modified By: Cyndie Cramer
# Description: This program evaluates housing data and
# calculates the Root Mean Squared Error (RMSE).
# =====================================================

# import numpy
import numpy as np
import pandas as pd   # Pandas is used here to read the data file

# Set the path to your data file
file_path = r"C:\Users\cyndi\Documents\week1data\housing_data.csv"


# Load the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)


# Convert the true and predicted price columns to NumPy
actual = df["sale_price"].to_numpy()
predicted = df["sale_price_pred"].to_numpy()

 
# RMSE function using NumPy
def rmse(y_true, y_pred):
    squared_diff = (y_true - y_pred)**2
    mean_squared_diff = np.mean(squared_diff)
    rmse_value = np.sqrt(mean_squared_diff)
    return rmse_value


# Compute RMSE
rmse_value = rmse(actual, predicted)

# Print result
print("RMSE:", rmse_value)
```
