```python
# ================================================================
# Title: Assignment 3.2 - Customer Spending Data
# Author: sajal2692 (GitHub username)
# Date: 12/21/2025
# Modified By: Cyndie Cramer
# Description: Creates a histogram of the customer spending 
# amounts, applies a log transformation to the data, and then 
# creates a histogram of the transformed data. This allows
# comparison of the original and log-transformed distributions, 
# which can help with modeling.
# =================================================================


# Import the libraries we need
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset using the full file path
data = pd.read_csv(r"C:\Users\cyndi\Documents\week3data\customer_spending.csv")

# Log transform the annual spending
data['LogAnnSpending'] = np.log(data['ann_spending'] + 1)

# Create side-by-side histograms
plt.figure(figsize=(14,5))

# Original spending
plt.subplot(1, 2, 1)
plt.hist(data['ann_spending'], bins=30, color='skyblue', edgecolor='black')
plt.title('Original Customer Annual Spending')
plt.xlabel('Annual Spending')
plt.ylabel('Frequency')

# Log-transformed spending
plt.subplot(1, 2, 2)
plt.hist(data['LogAnnSpending'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Log-Transformed Customer Annual Spending')
plt.xlabel('Log(Annual Spending + 1)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
