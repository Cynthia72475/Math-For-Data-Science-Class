```python
# =====================================================
# Title: Assignment 2.2 Question 6
# Author: sabnisshreyas91 (GitHub username)
# Date: 12/13/2025
# Modified By: Cyndie Cramer
# Description: Make a scatterplot of the highway miles 
# per gallon (y-axis) versus the weight (x-axis). Find
# the best-fit (least-squares) line using the 
# Scikit-learn package.
# =====================================================

# Import the libraries we need
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the dataset using the full file path
data = pd.read_csv(r"C:\Users\cyndi\Documents\week2data\car_data.csv")

# Prepare the variables
X = data["weight"].values.reshape(-1, 1)  # scikit-learn expects 2D array for features
y = data["hwy_mpg"].values

# Create the linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Get the slope (coefficient) and y-intercept
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope}")
print(f"Y-intercept: {intercept}")

# Predict y values using the model
y_pred = model.predict(X)

# Plot scatter plot
plt.scatter(data["weight"], data["hwy_mpg"], label="Data points")

# Plot the regression line
plt.plot(data["weight"], y_pred, color="red", label=f"y = {slope:.4f}x + {intercept:.2f}")

# Add labels and title
plt.xlabel("Vehicle Weight (pounds)")
plt.ylabel("Highway Miles per Gallon")
plt.title("Highway MPG vs Vehicle Weight with Best-Fit Line")
plt.legend()

# Show the plot
plt.show()
```
